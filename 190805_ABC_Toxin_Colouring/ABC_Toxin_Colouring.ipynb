```python
from ete3 import Tree
import colour_tree
import attributes

# Following commands will auto reload any changes to external modules
%load_ext autoreload
%autoreload 2
```

```python
# Define the tree to use     
tree = Tree('./Files/abc_toxins.nwk')

attribute_dict = attributes.attribute_dict

# Define the colours to use for each type - list of colours are here - http://etetoolkit.org/docs/latest/reference/reference_treeview.html
colour_dict = {'type1_single' :'dodgerblue', 'type2b_single': 'gold', 'type2a_single':'forestgreen', 'multiple' : 'red', 'unknown' : 'black'}
```

```python
# Save the coloured tips to file
output_tree, ts = colour_tree.colour_tips(tree, attribute_dict, colour_dict, outpath='./Output/coloured_tips.png', custom_layout=False)
```

```python
# Output the coloured tips image to the notebook
output_tree.render("%%inline", dpi=300, tree_style=ts)
```

```python
# Set the two attributes you wish to compare
first_attr = 'type2a_single'
second_attr = 'type2b_single'
```

```python
# Colour tree based on two attributes and propogate colours up as far as possible
output_tree, ts = colour_tree.colour_internal(tree, attribute_dict, colour_dict, first_attr, second_attr, outpath='./Output/coloured_nodes.png', custom_layout=False)
```

```python
# Colour tree based on two attibutes (display in Notebook)
output_tree.render("%%inline", dpi=300, tree_style=ts)
```
